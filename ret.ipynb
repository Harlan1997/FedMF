```python
arr = []
with open('ret.log', 'r') as f:
    lines = f.readlines()
    for line in lines:
        if line.startswith('loss '):
            arr.append(line[5:].strip('\n'))
```

```python
ret = list(map(lambda x : float(x), arr))
ret
```

```python
from matplotlib import pyplot as plt
import numpy as np
y = ret
x = np.arange(0, len(arr))
plt.title('FedCF training result')
plt.xlabel('iteration')
plt.ylabel('loss')
plt.plot(x, y)
plt.savefig('./ret.png')
plt.show()
```
